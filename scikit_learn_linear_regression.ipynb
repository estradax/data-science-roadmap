```python
import seaborn as sns

dataset = sns.load_dataset('tips')
dataset_sub = dataset.iloc[:120]
dataset_final = dataset_sub[["total_bill", "tip"]]

dataset_final.tail()
```

```python
from sklearn.linear_model import LinearRegression

X = dataset_final['total_bill'].to_numpy().reshape(-1, 1)
y = dataset_final['tip'].to_numpy().reshape(-1, 1)

reg = LinearRegression()
reg.fit(X, y)

reg.predict([[1]])
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return reg.predict([[x]])[0][0]
    
xs = np.array(range(120))
ys = np.array(list(map(f, xs)))

plt.xlim(right=60)
plt.scatter(dataset_final['total_bill'], dataset_final['tip'])
plt.plot(xs, ys)
```

```python
dataset_test = dataset.iloc[120:]

dataset_test.head()

xs = np.array(range(120))
ys = np.array(list(map(f, xs)))

plt.xlim(right=60)
plt.scatter(dataset_test['total_bill'], dataset_test['tip'])
plt.plot(xs, ys)
```
