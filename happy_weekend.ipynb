```python
import torch
import matplotlib.pyplot as plt

from torch import nn, optim

X = torch.tensor([0, 0.5, 1]).view(-1, 1).float()
y = torch.tensor([
    [1, 0],
    [0, 1],
    [1, 0],
]).float()

plt.scatter(X, torch.argmax(y, dim=1))
```

```python
torch.manual_seed(1)

class BadNet(nn.Module):
    def __init__(self, in_features, out_features):
        super().__init__()
        
        hidden = 4
        
        self.seq = nn.Sequential(
            nn.Linear(in_features, hidden),
            nn.ReLU(),
            nn.Linear(hidden, hidden),
            nn.ReLU(),
            nn.Linear(hidden, out_features),
        )
        
    def forward(self, x):
        return self.seq(x)
    
model = BadNet(1, 2)
criterion = nn.CrossEntropyLoss()
optimizer = optim.SGD(model.parameters(), lr=1e-1)
total_loss = []
```

```python
model.train()

for epoch in range(1000):
    optimizer.zero_grad()
    
    preds = model(X)
    loss = criterion(preds, y)
    
    loss.backward()
    optimizer.step()
    
    total_loss.append(loss.item())
    
plt.plot(total_loss)
```

```python
model.eval()
with torch.no_grad():
    preds = model(X)
        
plt.scatter(X, torch.argmax(y, dim=1))
plt.scatter(X, torch.argmax(preds, dim=1))
preds
```
