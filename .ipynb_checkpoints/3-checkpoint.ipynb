```python
import numpy as np
import matplotlib.pyplot as plt

X = np.array([np.arange(10), np.linspace(1, 5, 10)]).T
y = X[:,0]*2 + X[:,1]*3 + 5

ax = plt.axes(projection='3d')
ax.scatter3D(X[:,0], X[:,1], y)
```

```python
def parameters():
    w1 = np.random.rand()
    w2 = np.random.rand()
    b = 0
    return w1, w2, b

def loss(o, p):
    return (o - p)**2

def gradient_descent(o, p, w1, w2, b):
    w1 = -2 * (o - p) * w1
    w2 = -2 * (o - p) * w2
    b = -2 * (o - p) 
    return w1, w2, b

def step(w1, w2, b, sw1, sw2, sb):
    w1 = w1 - sw1
    w2 = w2 - sw2
    b = b - sb
    return w1, w2, b
```

```python
np.random.seed(42069)
w1, w2, b = parameters()
lr = 0.001
losses = []
for epoch in range(100):
    z_predicted = X[:,0]*w1 + X[:,1]*w2 + b
   
    err = loss(y, z_predicted)
    losses.append(np.sum(err) / len(err))
    
    dw1, dw2, db = gradient_descent(y, z_predicted, w1, w2, b)
    sw1 = np.sum(dw1) * lr
    sw2 = np.sum(dw2) * lr
    sb = np.sum(db) * lr
    
    w1, w2, b = step(w1, w2, b, sw1, sw2, sb)
    
print('w1:', w1, 'w2:', w2, 'bias:', b)
plt.plot(losses)
```

```python
X_train = np.arange(10)
Y_train = np.linspace(1, 8, 10)
z_train = X[:,0]*w1 + X[:,1]*w2 + b

z_test = X[:,0]*2 + X[:,1]*3 + 5

ax = plt.axes(projection='3d')
ax.plot3D(X_train, Y_train, z_train)
ax.scatter3D(X_train, Y_train, z_test)
```
