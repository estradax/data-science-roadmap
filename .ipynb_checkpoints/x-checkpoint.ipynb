```python
import numpy as np
import matplotlib.pyplot as plt

x = np.random.randn(20)
y = x*2 + np.random.rand()

plt.scatter(x, y)
```

```python
# f(x) = mx + b
def f(x, m, b):
    return m*x + b

f_vectorized = np.vectorize(f)

x_line = np.arange(1, np.max(x) + 1)
y_line = f_vectorized(x_line, 1, 0)

plt.scatter(x, y)
plt.plot(x_line, y_line)
```

```python
# resid = (actual_y - observed_y)^2
def resid(actual_y, observed_y):
    return (actual_y - observed_y)**2

resid_vectorized = np.vectorize(resid)

actual_y = y
observed_y = f_vectorized(x, 1, 0)
    
residuals = resid_vectorized(actual_y, observed_y)
np.sum(residuals)
```

```python
intercept_x = np.arange(0, 3, 0.1)
sum_of_squared_resid = []

for intercept in intercept_x:
    observed_y = f_vectorized(x, 1, intercept)
    
    residuals = resid_vectorized(actual_y, observed_y)
    sum_of_squared_resid.append(np.sum(residuals))
    
plt.scatter(intercept_x, sum_of_squared_resid)
```

```python
def deriv_intercept(observed_v, predicted_v):
    return -2 * (observed_v - predicted_v) * 1

deriv_intercept_vectorized = np.vectorize(deriv_intercept)

def deriv_slope(observed_v, predicted_v, slope):
    return -2 * (observed_v - predicted_v) * slope

deriv_slope_vectorized = np.vectorize(deriv_slope)

intercept = 0
slope = 0.25

losses = []

for epoch in range(800):
    y_predicted = f_vectorized(x, slope, intercept)
    
    resid = resid_vectorized(y, y_predicted)
    loss = np.sum(resid)
    losses.append(loss)
    
    step_size_intercept = np.sum(deriv_intercept_vectorized(y, y_predicted)) * 0.01
    intercept = intercept - step_size_intercept
    
    step_size_slope = np.sum(deriv_slope_vectorized(y, y_predicted, slope)) * 0.01
    slope = slope - step_size_slope

plt.plot(losses, np.arange(len(losses)))    
```
