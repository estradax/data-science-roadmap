```python
import numpy as np
import matplotlib.pyplot as plt

X_train = np.arange(10)
y_train = X_train*2 + 5

plt.scatter(X_train, y_train)
```

```python
def parameters():
    w = np.random.rand()
    b = np.random.rand()
    return w, b

def loss(o, p):
    return (o - p)**2

def gradient_descent(o, p, w, b):
    return (-2 * (o - p) * w), (-2 * (o - p))

def step(w, b, sw, sb):
    return w - sw, b - sb
```

```python
np.random.seed(42069)
w, b = parameters()
lr = 0.001
for epoch in range(40):
    y_predicted = X_train*w + b
   
    err = loss(y_train, y_predicted)
    print('loss:', np.sum(err) / len(err))
    
    dw, db = gradient_descent(y_train, y_predicted, w, b)
    sw = np.sum(dw) * lr
    sb = np.sum(db) * lr
    
    w, b = step(w, b, sw, sb)
    
print('weight:', w, 'bias:', b)
```

```python
x = np.arange(10)
y = x*w + b

plt.scatter(X_train, y_train)
plt.plot(x, y)
```
