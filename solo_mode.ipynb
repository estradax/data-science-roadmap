```python
def show_progress(xs, ys):
    dataset_final.plot.scatter(x="total_bill", y="tip")
    plt.plot(xs, ys)
```

```python
epochs = 20
learning_rate = 0.01

xs = np.array(range(size))
ys = np.array([])

rs = np.array([])

for epoch in range(epochs):
    for x in xs:
        m = epoch * learning_rate
        y = m*x + 0
        
        ys = np.append(ys, y)

    r_sum = 0
    for i, y in enumerate(ys):
        r = (y - dataset_final['tip'][i])**2
        r_sum += r
        
    show_progress(xs, ys)
        
    rs = np.append(rs, r_sum)    
    ys = np.array([])

plt.show()
```
