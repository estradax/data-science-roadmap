```python
import numpy as np
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
import torch.nn.functional as F

X = np.linspace(0, 1, 10, dtype='float32')
y = X * 2 + 5

plt.scatter(X, y)
torch.tensor(X[0])
```

```python
torch.random.manual_seed(69420)

class LinearRegression(nn.Module):
    def __init__(self):
        super().__init__()
        self.linear = nn.Linear(1, 1)
        
    def forward(self, x):
        return self.linear(x)
 
model = LinearRegression()
```

```python
loss_fn = nn.MSELoss()
optimizer = torch.optim.SGD(model.parameters(), lr=0.1)

losses = []

model.train()
for epoch in range(10):
    total_loss = 0
    for i in range(len(X)):
        
        pred = model(torch.tensor([X[i]]))
        loss = loss_fn(pred, torch.tensor([y[i]]))
        total_loss += float(loss)
        
        loss.backward()
        
    losses.append(total_loss)
        
    optimizer.step()
    optimizer.zero_grad()
    
plt.plot(losses)
```

```python
with torch.no_grad():
    plt.scatter(X, y)
    y = []
    for x in X:
        y.append(model(torch.tensor([x])))
    plt.plot(X, y)
```
