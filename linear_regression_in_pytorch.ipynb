```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42069)

X = np.arange(20)
y = X**2 + np.random.rand()

plt.scatter(X, y)
```

```python
import torch
import torch.nn as nn

torch.random.manual_seed(69420)

class LinearRegression(nn.Module):
    def __init__(self):
        super().__init__()
        self.w = nn.Parameter(torch.rand(1))
        self.b = nn.Parameter(torch.rand(1))
        
        self.double()
        
    def forward(self, x):
        out = x*self.w + self.b
        out = torch.relu(out)
        return out
 
model = LinearRegression()
```

```python
loss_fn = nn.MSELoss()
optimizer = torch.optim.SGD(model.parameters(), lr=0.001)

model.train()
for epoch in range(10):
    pred = model(torch.from_numpy(X))
    loss = loss_fn(pred, torch.from_numpy(y))
        
    loss.backward()
    optimizer.step()
    optimizer.zero_grad()
    
```
